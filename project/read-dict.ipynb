```python
f = open('sp500names.txt', 'r')
g = open('orgs.perline.txt', 'r')
h = open('sp500.orgs.txt', 'w')

orgs = []

for gg in g:
    orgs.append(gg)
g.close()

sp500n = []

for ff in f:
    sp500n.append(ff)
f.close()

i = 0
for s in sp500n:
    first = s.split()[0]
    i = i + 1
    for org in orgs:
        if first in org:
            h.write(str(i) + '|' + first + '|' + org)

h.close()

            
    
    
```
