```python
import os
import xml.etree.ElementTree as ET
from sets import Set
s= Set([])

f = open('dict-87to91.db', 'w')

# Set the directory you want to start from
rootDir = 'data'
for dirName, subdirList, fileList in os.walk(rootDir):
    for fname in fileList:
        dict = {}
        dict['guid'] = fname[:-4]
        dict['pdate'] = dirName[5:]
        dict['org'] = ''
        
        tree = ET.parse(dirName + '/' + fname)
        for meta in tree.getroot().iter('meta'):
            if not meta.get('name') == '':
                dict[meta.get('name')] = meta.get('content')
        for classifier in tree.getroot().iter('classifier'):
            if not classifier.get('type') == '':
                dict[classifier.get('type')] = classifier.text
        for byline in tree.getroot().iter('byline'):
            if byline.get('class') == 'normalized_byline':
                dict['normalized_byline'] = byline.text
        for org in tree.getroot().iter('org'):
            if org.text is not None:
                dict['org'] = dict['org'] + ', ' + org.text

        dict['org'] = dict['org'][2:]
        f.write(str(dict)+',\n')
f.write('{}\n')
f.close()
```
